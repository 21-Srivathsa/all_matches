```python
def player_matchups(a,b):
    runs,wickets = [],[]
    for i in range(len(data["match_id"])):
        if data["striker"][i] == a and data["bowler"][i] == b:
            runs.append(data["runs_off_bat"][i])
            if data["wicket_type"][i] == "bowled":
                wickets.append("bowled")
            elif data["wicket_type"][i] == "caught":
                wickets.append("caught")
            elif data["wicket_type"][i] == "lbw":
                wickets.append("lbw")
            elif data["wicket_type"][i] == "stumped":
                wickets.append("stumped")
    if len(runs) == 0:
        return "No faceoff between them or incorrect name entered"
    fours,sixes,sr,wick = runs.count(4),runs.count(6),sum(runs)*100/len(runs),len(wickets)
    print("===MATCHUP BETWEEN ", a, " AND ",b,"===")
    print(a, " Has scored ",sum(runs)," in",len(runs)," balls with a strike rate of ",sum(runs)*100/len(runs)," against ", b)
    print(a," hit ",runs.count(4)," fours and ",runs.count(6)," sixes against ",b)
    print("Number of times ",b," got ",a," out is ",len(wickets))
    for i in ["bowled","caught","stumped","lbw"]:
        print(wickets.count(i),i)
    return [fours,sixes,sr,wick]
```

```python
def points(a,b):
    point = 0
    values = player_matchups(a,b)
    if values == "No faceoff between them or incorrect name entered":
        print("No faceoff between ",a," and ",b)
        point = 0
    else:
        if values[2] >= 100 and values[2] < 150:
            point = 5*values[0] + 10*values[1] + 10 -20*values[3]
        if values[2] >= 150 and values[2] < 200:
            point = 5*values[0] + 10*values[1] + 15 -20*values[3]
        if values[2] >= 200:
            point = 5*values[0] + 10*values[1] + 20 -20*values[3]
        if values[2] < 100:
            point = 5*values[0] + 10*values[1] -20*values[3]
    return point
```

```python
def core(team1,team2):
    count1,count2 = 0,0
    for i in team1:
        for j in team2:
            count1 +=points(i,j)
            print("Points earned till now is ",count1)
    for i in team1:
        for j in team2:
            count2 +=points(j,i)
            print("Points earned till now is ",count2)
    count = count1+count2
    if count1-count2 > 0:
        print("Team 1 won with ", count1-count2," points!!!!!!")
    else:
        print("Team 2 won with ",abs(count1-count2)," points!!!!!!")
            
            
```

```python
def pickAI():
    ai = []
    for i in range(5):
        ai.append(random.choice(players))
    if len(set(ai)) != len(ai):
        ai = []
        for i in range(5):
            ai.append(random.choice(players))
    return ai
```
